# The Computational Graph
Every time you perform an operation on tensors, PyTorch builds a Dynamic Computational Graph (DCG).

* Nodes: These represent tensors (data).

* Edges: These represent functions or operations (addition, multiplication, etc.).

**Direction**: The graph is a Directed Acyclic Graph (DAG). It flows forward during the "Forward Pass" to calculate the output, and PyTorch "remembers" the path to go backward during the "Backward Pass."

Unlike some other frameworks that use "static" graphs (where the graph is defined once and then run), PyTorch uses **dynamic** graphs. The graph is recreated from scratch every time you run a forward pass, allowing you to use Python control flow like if statements or `for` loops in your model.

```python
import torch

a = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(3.0, requires_grad=True)
c = torch.tensor(10.0)

# The graph is built here
d = a + b
z = d * c

# The backward pass
z.backward()

# Gradients are now stored in a.grad and b.grad
print(a.grad) # Output: 10.0
```

```python
import torch
from torchviz import make_dot

a = torch.tensor(2.0, requires_grad=True)
b = torch.tensor(3.0, requires_grad=True)
c = torch.tensor(10.0)

d = a + b
z = d * c

# Visualize the graph
make_dot(z, params={'a': a, 'b': b}).render("computation_graph", format="png")
```
