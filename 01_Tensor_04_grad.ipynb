# Autograd

A simple graph
z = x * y , where y a + b . so three parameters

```python
import torch
```

```python
x = torch.tensor([2.0,1.5], requires_grad=True)
a = torch.tensor([5.0, 7.3], requires_grad=True)
b = torch.tensor([3.0,2.5], requires_grad=True)

print(x)
```

```python
y = x * a
z = y + b
```

```python
print(y.grad_fn)
print(y.grad)
```
